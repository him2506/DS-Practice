###### 
In the above plots, we can see that the points are overlapping each other, to eliminate this situation, we can set kind = â€œswarmâ€, swarm uses an algorithm that prevents the points from overlapping and adjusts the points along the categorical axis. Letâ€™s see how it looks like

```python
sns.catplot(x="number_of_doors", y="horsepower", hue="fuel_type",kind="swarm",data=auto)
```

```python
sns.catplot(x="number_of_doors", y="horsepower", kind = "box", data=auto)
```

##### Violin Plot using seaborn
We can also represent the above variables differently by using violin plots. Letâ€™s try it out

```python
sns.catplot(x="number_of_doors", y="horsepower", kind = "violin", data=auto)
```

```python
sns.catplot(x="number_of_doors", y="horsepower", kind = "violin",hue="fuel_type" ,data=auto)
```

##### 
The violin plots combine the boxplot and kernel density estimation procedure to provide richer description of the distribution of values. The quartile values are displayed inside the violin. We can also split the violin when the hue semantic parameter has only two levels, which could also be helpful in saving space on the plot. Letâ€™s look at the violin plot with a split of levels.

```python
sns.catplot(x="number_of_doors", y="horsepower", kind = "violin",hue="fuel_type",split=True,data=auto)
```

```python
sns.catplot(x="number_of_doors", y="horsepower", kind = "bar",hue="fuel_type",data=auto)
```

###### Pointplot using seaborn
Another type of plot coming in is pointplot, and this plot points out the estimate value and confidence interval. Pointplot connects data from the same hue category. This helps in identifying how the relationship is changing in a particular hue category. You can check out how does a pointplot displays the information below.

```python
sns.catplot(x="number_of_doors", y="horsepower", kind = "point",hue="fuel_type",data=auto)
```

```python
sns.catplot(x="number_of_doors", y="horsepower",hue="fuel_type",col="engine_location", aspect=.9,
            kind="swarm",data= auto)

#here we get sworm plot for both engine loaction
```
